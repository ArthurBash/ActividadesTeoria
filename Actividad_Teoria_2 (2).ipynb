```python
#Importo las librerias necesarias,el csv, y les modifico las indices de las columnas para mas legibilidad
import numpy as np
import matplotlib.pyplot as plot
import pandas as pd
file = pd.read_csv('Jugadas.csv')
file.columns = ["Tiempo", "partida", "Cantidad_palabras", "Evento","USR_nick", "USR_genero", "USR_edad", "Estado", "Palabra", "Nivel_juego"]

#file.columns

```

```python
# Aca filtro lo que voy a mostrar y armo la tabla de estadisticas

file['Estado'].fillna('nuevo',inplace = True)
filtro_Partidas_empezadas = file['Estado'] == 'nuevo'
filtro_a_mostrar = file[file['Estado'] == 'nuevo']['USR_nick'].value_counts().sort_values()
x = list(filtro_a_mostrar.index)
y = list(filtro_a_mostrar.values)
fig,ax = plot.subplots()
plot.xlabel('Jugadores')
plot.ylabel('Partidas jugadas')
plot.title('Cantidad de partidas jugadas por jugador')
ax.bar(x,y,width = 0.8,color = 'green')
plot.show()
```

```python
from PIL import Image
from wordcloud import WordCloud, STOPWORDS

animales = pd.read_csv('zoo.csv')

#Creo que filtro  de animales con plumas y pejaje, luego armo un texto con el resultado
filtro =  (animales['pelaje'] == '1') | (animales['plumas'] == '1')
animales_filtro =  list(animales[filtro]['animal_mame'].values)
text = str([c.replace(' ','_') for c in animales_filtro])



```

```python
#Defino la figura ,abro la imagen y creo el plot
def plot_cloud(wordcloud):
  plot.figure(figsize=(40, 30))
  plot.imshow(wordcloud)
  plot.axis("off");

mask = np.array(Image.open('v.jpeg'))
wordcloud = WordCloud(width = 300, height = 200, random_state=1,background_color='black', colormap='Set2', collocations=False, stopwords =STOPWORDS, mask=mask).generate(text)

plot_cloud(wordcloud)


```
